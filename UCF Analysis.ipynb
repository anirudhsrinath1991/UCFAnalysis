```python
#Import library and read the file
import pandas as pd
df = pd.read_excel('Excel1.xlsm')
```

```python
#Display the first 5 rows
df.head()
```

```python
#Display all the datatypes of the columns
df.dtypes
```

```python
#Create a New Variable and filter only the Four-Seam Fastballs in it
seamfastball_data = df[df['Pitch Type'] == 'Four Seam Fastball']
```

```python
#Display the head of the Four-Seam Fastballs
seamfastball_data.head()
```

```python
#Display the Pitch Speed of seam Fastball in a different variable
FourseamFastball = seamfastball_data['Pitch Speed [mph]']
```

```python
#Display the Stats of Pitch Speed of Seam Fastball
FourseamFastball.describe()
```

```python
#Therefore 66% of 4 Seam Fastballs falls between 85 to 91 mph. Creating a new variable called Two Seam Fastball to understand
twoseamfastball_data = df[df['Pitch Type'] == 'Two Seam Fastball']
```

```python
#Display the first 5 rows of Two Seam Fastballs.
twoseamfastball_data.head()
```

```python
#Display the spread of Two Seam Fastballs(Pitch Speed)
twoseamfastball_data['Pitch Speed [mph]'].describe()
```

```python
#66 % of Two Seam Fastballs pitch speeds fall between 87 to 89 mph.
#Find the total no of Pitch Types that are present(unique)
df['Pitch Type'].unique()
```

```python
#Filter Sliders 
slider_data = df[df['Pitch Type'] == 'Slider']
```

```python
#Display the spread of Slider Data(Pitch Speed)
slider_data['Pitch Speed [mph]'].describe()
```

```python
#66% of Slider Pitcher Speeds fall between 79 to 83 mph.
#Filter Curveballs
curveball_data = df[df['Pitch Type'] == 'Curveball']
```

```python
#Display the spread of Curveballs(Pitch Speeds)
curveball_data['Pitch Speed [mph]'].describe()
```

```python
#66% of Pitch Speeds(Mph) falls between 73 to 79 mph.
#Filter Cutter Data 
Cutter_data = df[df['Pitch Type'] == 'Cutter']
```

```python
#Display the spread of Cutters 
Cutter_data['Pitch Speed [mph]'].describe()
#66 % of Cutters Falls between 78 to 86 mph.
```

```python
#Filter CHangeUps
ChangeUp = df[df['Pitch Type'] == 'Changeup']
```

```python
#Stats of ChangeUps
ChangeUp['Pitch Speed [mph]'].describe()
```

```python
#Graph for Pitch Release Height vs Pitch Extension for All Pitch Types in UCF Dataset.
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
plt.figure(figsize=(14,10))
ax=plt.gca()
ax.set_xlim([0,10])
ax.set_ylim([0,10])
plt.xlabel('Pitch Extension [ft] (Fastball)')
plt.ylabel('Pitch Release Height [ft] (Fastball)')


x1 = seamfastball_data['Pitch Extension [ft]']
y1 = seamfastball_data['Pitch Release Height [ft]']
x2 = twoseamfastball_data['Pitch Extension [ft]']
y2 = twoseamfastball_data['Pitch Release Height [ft]']
x3 = Cutter_data['Pitch Extension [ft]']
y3 = Cutter_data['Pitch Release Height [ft]']
x4 = ChangeUp['Pitch Extension [ft]']
y4 = ChangeUp['Pitch Release Height [ft]']
plt.plot(x1,y1,'ro',label = 'Four-Seam')
plt.plot(x2,y2,'go', label = 'Two-Seam')
plt.plot(x3,y3,'bo', label = 'Cutterdata')
plt.plot(x4,y4,'yo', label = 'ChangeUp')
pylab.legend(loc='upper right')
```

```python
# Use Seaborne Library for Data Visualization
import seaborn as sns
sns.factorplot(data = seamfastball_data, x = 'Pitch Release Height [ft]', y = 'Pitch Extension [ft]')
plt.show()
```

```python
#Use Seaborne Library for Data Visualization(For each Pitch Type)
sns.lmplot(x = 'Pitch Release Height [ft]', y ='Pitch Extension [ft]', data = seamfastball_data)
sns.lmplot(x = 'Pitch Release Height [ft]', y ='Pitch Extension [ft]', data = twoseamfastball_data)
sns.lmplot(x = 'Pitch Release Height [ft]', y ='Pitch Extension [ft]', data = curveball_data)
sns.lmplot(x = 'Pitch Release Height [ft]', y ='Pitch Extension [ft]', data = Cutter_data)
```
